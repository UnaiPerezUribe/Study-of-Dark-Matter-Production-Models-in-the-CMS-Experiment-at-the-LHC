```python
## total

import numpy as np
import matplotlib.pyplot as plt

archivo_total1 = "todo_xsec.txt"


columna1 = []
columna2 = []


with open(archivo_total1, "r") as file:
    for linea in file:
        valores = linea.strip().split()  
        if len(valores) > 1:  
            try:
                columna1.append(float(valores[0]))  
                columna2.append(float(valores[1]))  
            except ValueError:
                print("Línea inválida:", linea)


x_total1 = np.array(columna1)
xsec_total1 = np.array(columna2)


archivo_total2 = "todo_xsec2.txt"


columna1 = []
columna2 = []


with open(archivo_total2, "r") as file:
    for linea in file:
        valores = linea.strip().split()  
        if len(valores) > 1:  
            try:
                columna1.append(float(valores[0])) 
                columna2.append(float(valores[1])) 
            except ValueError:
                print("Línea inválida:", linea)


x_total2 = np.array(columna1)
xsec_total2 = np.array(columna2)

x_total = np.concatenate((x_total1, x_total2))
xsec_total = np.concatenate((xsec_total1, xsec_total2))


plt.plot(x_total, xsec_total, marker="o", linestyle="-", color="b", label="Datos del archivo")
plt.xlabel("Valores de la primera columna")
plt.ylabel("Valores de la segunda columna")
plt.title("Gráfico de bb_xsec.txt")
plt.legend()


plt.show()
```
