```python
def DrawOverflow(h):
    # This function paints the histogram h with an extra bin for overflows
    nx = h.GetNbinsX() + 1
    xbins = np.zeros(nx + 1)
    
    for i in range(nx):
        xbins[i] = h.GetBinLowEdge(i + 1)
    
    xbins[nx] = xbins[nx - 1] + h.GetBinWidth(nx)
    
    temp_name = f"{h.GetName()}wtOverFlow"
    htmp = ROOT.TH1F(temp_name, h.GetTitle(), nx, xbins)
    
    # Style
    htmp.SetLineColor(h.GetLineColor())
    htmp.SetLineWidth(h.GetLineWidth())
    htmp.SetLineStyle(h.GetLineStyle())
    
    # Reset axis labels
    htmp.SetXTitle(h.GetXaxis().GetTitle())
    htmp.SetYTitle(h.GetYaxis().GetTitle())
    
    # Fill new histogram including extra bin for overflows
    for i in range(1, nx + 1):
        htmp.Fill(htmp.GetBinCenter(i), h.GetBinContent(i))
    
    # Fill underflows
    htmp.Fill(h.GetBinLowEdge(1) - 1, h.GetBinContent(0))
    
    # Restore number of entries
    htmp.SetEntries(h.GetEntries())
    
    # Fill style and color
    htmp.SetFillStyle(h.GetFillStyle())
    htmp.SetFillColor(h.GetFillColor())
    
    return htmp
```

```python
# Distribución Zp s alto con fondos

import ROOT
import random


# Estilo ROOT
ROOT.gROOT.ProcessLine(".L /eos/home-u/uperezur/LatinoTreesLHE/TDRStyle.cc")
ROOT.TDRStyle()
ROOT.gStyle.SetTitleYOffset(1.1)
ROOT.gStyle.cd()

# === Variables configurables ===
x_min = 0
x_max = 8
points = 50
xsec = 0 # 1 escalado por xsec y 0 no 
variable = "dRl1l2"
nombre = "#it{mll} (GeV)"

# === Definir función de cortes ===
def pasa_cortes(entry):
                                                                                
    #return( entry.ptl1 > 45 and entry.ptl2 > 10 and entry.mll > 110 and entry.met > 60 and entry.ptll > 35
        # and entry.mtllmet > 75 and entry.dRl1l2 < 2.7 and entry.dRl1l2 > 1
           #and entry.dRllmet > 3.1 and entry.dRl1met > 2.8 and entry.dRl1met > 2.2)
    
    return entry.ptl1 > 0



# === Listas de árboles, títulos y colores ===
trees = [t_1_Zp_s300, t_3_Zp_s300, t_5_Zp_s300, t_7_Zp_s300, t_9_Zp_s300, t_11_Zp_s300, t_13_Zp_s300,
         t_1_WW, t_1_t, t_1_hb]
         #t_WW_total, t_tt_total, t_1_hb]

titles = [
    "Zp 195 GeV", "Zp 200 GeV", "Zp 295 GeV", "Zp 300 GeV",
    "Zp 500 GeV", "Zp 1000 GeV", "Zp 1500 GeV",
    "WW", "t#bar{t}", "h #rightarrow WW"
]

colors = [
    ROOT.kPink+6, ROOT.kRed, ROOT.kGreen, ROOT.kOrange+7,
    ROOT.kMagenta+2, ROOT.kCyan+1, ROOT.kYellow,
    ROOT.kBlue, ROOT.kViolet+1, ROOT.kAzure+7
]

# === Crear histogramas ===
histos = []
for i, title in enumerate(titles):
    h = ROOT.TH1F(f"h_{i+1}", title, points, x_min, x_max)
    h.Sumw2()
    h.SetMarkerSize(0)
    h.SetMarkerColor(colors[i])
    h.SetLineColor(colors[i])
    h.SetLineWidth(3)
    if i == 0:
        h.GetXaxis().SetTitle(nombre)
    if i >= 7:  # Fondos
        h.SetFillColorAlpha(colors[i], 0.3 if i != 7 else 0.1)
    histos.append(h)

# === Llenar los histogramas ===
for i, (tree, hist) in enumerate(zip(trees, histos)):
    for entry in tree:
        if pasa_cortes(entry):
            value = getattr(entry, variable)
            if xsec == 1:
                if i in [0, 1, 2, 3, 4, 5, 6]:  # señales Zp bajas
                    weight = entry.evtWeight
                elif i == 8:  # t#bar{t}
                    weight = entry.evtWeight * 0.01
                else:
                    weight = entry.evtWeight
                hist.Fill(value, weight)
            if xsec == 0:
                hist.Fill(value)
            if xsec == 2:
                if i in [0, 1, 2, 3, 4, 5, 6]:  # señales Zp bajas
                    weight = entry.evtWeight*200
                elif i == 8:  # t#bar{t}
                    weight = entry.evtWeight * 0.01
                else:
                    weight = entry.evtWeight
                hist.Fill(value, weight)

# === Crear el canvas ===
n = random.randint(0, 500)
canvas = ROOT.TCanvas(f"cn_{n}", f"cn_{n}", 800, 600)

# === Crear el stack con los fondos ===
stack = ROOT.THStack("stack", "")
for h in histos[7:]:
    stack.Add(h)

# === Dibujar el stack ===
stack.Draw("HIST")
stack.GetXaxis().SetTitle(nombre)
stack.GetYaxis().SetTitle("Eventos")
stack.SetMaximum(1.2 * max(stack.GetMaximum(), max(h.GetMaximum() for h in histos[:7])))

# === Dibujar las señales encima ===
for h in histos[:7]:
    h.SetLineStyle(1)
    h.Draw("HIST SAME")

# === Crear y dibujar la leyenda ===
leg = ROOT.TLegend(0.7, 0.9, 0.9, 0.5)
for i, h in enumerate(histos):
    draw_opt = "f" if i >= 7 else "l"
    leg.AddEntry(h, h.GetTitle(), draw_opt)
leg.SetBorderSize(0)
leg.SetFillStyle(0)
leg.SetTextSize(0.035)
leg.Draw()

# === Mostrar el canvas ===

canvas.Draw()


seniales = {
    "195 GeV": t_1_Zp_s300,
    "200 GeV": t_3_Zp_s300,
    "295 GeV": t_5_Zp_s300,
    "300 GeV": t_7_Zp_s300,
    "500 GeV": t_9_Zp_s300,
    "1000 GeV": t_11_Zp_s300,
    "1500 GeV": t_13_Zp_s300
}

# Árboles de fondo
fondos = {
    "WW": t_1_WW,
    "h→WW": t_1_hb,
    "tt̄": t_1_t
}

from prettytable import PrettyTable
import math

def contar_eventos(tree, usar_pesos=False, escala=1.0):
    total = 0
    for entry in tree:
        if pasa_cortes(entry):
            if usar_pesos:
                total += entry.evtWeight * escala
            else:
                total += 1
    return total
usar_pesos = (xsec == 1)

tabla = PrettyTable()
tabla.field_names = ["Masa señal", "Eventos señal", "WW", "h→WW", "tt̄", "Eficacia", "Error"]

for masa, tree_s in seniales.items():
    n_s = contar_eventos(tree_s, usar_pesos)

    n_WW = contar_eventos(fondos["WW"], usar_pesos)
    n_HWW = contar_eventos(fondos["h→WW"], usar_pesos)
    n_tt = contar_eventos(fondos["tt̄"], usar_pesos, escala=0.01 if usar_pesos else 1.0)

    n_total = n_s + n_WW + n_HWW + n_tt
    eficacia = n_s / n_total if n_total > 0 else 0
    error = math.sqrt(n_s) / n_total if n_total > 0 else 0

    # Mostrar enteros si no se usan pesos, decimales si sí
    if not usar_pesos:
        row = [masa, int(n_s), int(n_WW), int(n_HWW), int(n_tt), f"{eficacia:.3f}", f"{error:.3f}"]
    else:
        row = [masa, f"{n_s:.1f}", f"{n_WW:.1f}", f"{n_HWW:.1f}", f"{n_tt:.1f}", f"{eficacia:.3f}", f"{error:.3f}"]

    tabla.add_row(row)

print(tabla)
```

```python
# Zp asintomatic

import ROOT
import math

ROOT.gROOT.ProcessLine(".L /eos/home-u/uperezur/LatinoTreesLHE/TDRStyle.cc")
ROOT.TDRStyle()
ROOT.gStyle.SetTitleYOffset(1.1)
ROOT.gStyle.cd()

# variables a modificar
x_min = 0
x_max = 8
points = 50
xsec = 0 # 1 escalado por xsec y 0 no
s_alto = 0 # 1 si, 0 no

variable = "dRl1l2"
nombre = "#it{Mt_{ll}} (GeV)"

def pasa_cortes(entry):
    return( entry.ptl1 > 55 and entry.ptl2 > 24 and entry.met > 80 and entry.ptll > 80 and entry.mtllmet > 162 
           and entry.mtl1met > 132 and entry.mtl2met > 102 and entry.dRl1l2 < 2)
    return entry.ptl1 > 0

# Árboles de señal y fondo
if s_alto == 0:
    signals = [t_1_Zp, t_3_Zp, t_5_Zp, t_7_Zp, t_9_Zp, t_11_Zp, t_13_Zp]
if s_alto == 1:
    signals = [t_1_Zp_s300, t_3_Zp_s300, t_5_Zp_s300, t_7_Zp_s300, t_9_Zp_s300, t_11_Zp_s300, t_13_Zp_s300]

signal_titles = ["Zp 195 GeV", "Zp 200 GeV", "Zp 295 GeV", "Zp 300 GeV",
    "Zp 500 GeV", "Zp 1000 GeV", "Zp 1500 GeV"]
backgrounds = [t_1_WW, t_1_t, t_1_hb]

colors = [ROOT.kPink+6, ROOT.kRed, ROOT.kGreen, ROOT.kOrange+7,
    ROOT.kMagenta+2, ROOT.kCyan+1, ROOT.kYellow]

# Histogramas de señal
signal_histos = []
for i, (sig_tree, title, color) in enumerate(zip(signals, signal_titles, colors)):
    h = ROOT.TH1F(f"h_sig_{i}", title, points, x_min, x_max)
    h.Sumw2()
    for entry in sig_tree:
        if pasa_cortes(entry):
            value = getattr(entry, variable)
            if xsec == 0:
                h.Fill(value) 
            if xsec == 1:
                h.Fill(value, weight)
            if xsec == 2:
                h.Fill(value, weight*200)
    signal_histos.append(h)

# Histograma combinado de fondo
h_bkg = ROOT.TH1F("h_bkg", "Background", points, x_min, x_max)
h_bkg.Sumw2()
for bkg_tree in backgrounds:
    for entry in bkg_tree:
        if pasa_cortes(entry):
            value = getattr(entry, variable)
            if xsec == 0:
                h_bkg.Fill(value)
            if xsec == 1:
                if bkg_tree == t_1_t:
                    weight = entry.evtWeight * 0.01  # Reescala ttbar
                else:
                    weight = entry.evtWeight
                h_bkg.Fill(value, weight)
            if xsec == 2:
                if bkg_tree == t_1_t:
                    weight = entry.evtWeight * 0.01  # Reescala ttbar
                else:
                    weight = entry.evtWeight
                h_bkg.Fill(value, weight)

# Histogramas de significancia asintótica
asymptotic_significance_histos = []
for i, (h_sig, color) in enumerate(zip(signal_histos, colors)):
    h_asym = ROOT.TH1F(f"h_asymptotic_{i}", f"Asintótica {signal_titles[i]}", points, x_min, x_max)
    h_asym.Sumw2()
    h_asym.SetLineColor(color)
    h_asym.SetLineWidth(2)
    h_asym.SetMarkerSize(0)

    for bin in range(1, h_asym.GetNbinsX() + 1):
        S = h_sig.GetBinContent(bin)
        B = h_bkg.GetBinContent(bin)
        if B > 0 and S > 0:
            q0A = 2 * ((S + B) * math.log(1 + S / B) - S)
            significance = math.sqrt(q0A)
        else:
            significance = 0
        h_asym.SetBinContent(bin, significance)

    asymptotic_significance_histos.append(h_asym)

# Dibujar
leg = ROOT.TLegend(0.75, 0.88, 0.93, 0.6)
leg.SetBorderSize(0)
leg.SetFillStyle(0)
leg.SetTextSize(0.04)

cn_asym = ROOT.TCanvas("cn_asym", "Significancia asintótica", 800, 600)

for i, h_asym in enumerate(asymptotic_significance_histos):
    h_asym.GetXaxis().SetTitle(nombre)
    h_asym.GetYaxis().SetTitle("#sqrt{q_{0,A}}")
    if i == 0:
        h_asym.SetMaximum(max(h.GetMaximum() for h in asymptotic_significance_histos) * 1.2)
        h_asym.Draw("hist")
    else:
        h_asym.Draw("histsame")
    leg.AddEntry(h_asym, h_asym.GetTitle(), "l")

leg.Draw()
cn_asym.Draw()

print("Máximo de √q0,A para cada señal:")
for h_asym, title in zip(asymptotic_significance_histos, signal_titles):
    max_bin = h_asym.GetMaximumBin()
    max_val = h_asym.GetBinContent(max_bin)
    x_val = h_asym.GetXaxis().GetBinCenter(max_bin)
    print(f"  {title}: máximo = {max_val:.3f} en x = {x_val:.1f}")
```
